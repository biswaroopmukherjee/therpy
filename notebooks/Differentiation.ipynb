```python
import therpy as tp
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

import zsharp as zs
```

# Uniform Grid derivatives

```python
x = np.linspace(0,5,100)

# define a signal
freq=2
signal_strength=1
signal = signal_strength*np.sin(freq*x)
deriv = signal_strength*np.cos(freq*x)

noise_strength = 0.1
noise = noise_strength*np.random.randn(len(x))
```

```python
# plt.plot(x,signal, 'grey')
plt.plot(x,signal+noise)
plt.show()
```

```python
plt.plot(x, tp.savitzky_golay(signal+noise,11,5,1))
plt.plot(x, deriv)
```
