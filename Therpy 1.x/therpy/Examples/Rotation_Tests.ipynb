```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import scipy.ndimage.interpolation
```

```python
def imshow(data, ax=None):
    if ax is None:
        fig, ax = plt.subplots()
    ax.imshow(data, interpolation='nearest', cmap='viridis',)
    
def imrotate(data, angle):
    return scipy.ndimage.interpolation.rotate(data, angle, reshape=False, order=1, mode='nearest', cval=0.0, prefilter=False)
```

```python
# Test matrix
m = np.zeros(shape=(50,50))
m[:,25] = 1

m1 = m.copy()
for i in range(45):
    m1 = imrotate(m1, 1)
m2 = imrotate(m, 45)

plt.figure()
plt.imshow(m1, interpolation='nearest', cmap='viridis',)
plt.colorbar()

plt.figure()
plt.imshow(m2, interpolation='nearest', cmap='viridis',)
plt.colorbar()
```
