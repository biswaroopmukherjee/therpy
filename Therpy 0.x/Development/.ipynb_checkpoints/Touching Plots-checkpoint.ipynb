```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x = np.arange(6)
data = np.array([np.random.normal(size=50) for i in x])
```

```python
L = len(x)
fig, ax = plt.subplots(ncols=L, nrows=1, figsize=[2*L, 3], sharey=True)
for i in range(len(x)):
    ax[i].plot(data[i], 'C{}'.format(i%10))
plt.subplots_adjust(wspace=0.05, hspace=0)

```
