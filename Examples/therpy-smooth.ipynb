# Binning by x

````python
# Generate data
x1 = np.linspace(0,4,100)
y1 = -np.sin(x1) + np.random.normal(size=x1.shape,scale=0.2)
x2 = np.linspace(0,4**0.5,100)**2
y2 = np.sin(x2**2) + np.random.normal(size=x2.shape,scale=0.2)
plt.plot(x1,y1,'b.')
plt.plot(x2,y2,'r.')
plt.title('blue: equal spacing & red: unequal spacing')
````

### 10 bins with and without bounds

````python
x1_binned, y1_binned = tp.smooth.binbyx(x1,y1,bins = 20)
x2_binned, y2_binned = tp.smooth.binbyx(x2,y2, bins = 20, sects = [0.5,2])
plt.plot(x1_binned, y1_binned,'b.-')
plt.plot(x2_binned, y2_binned,'r.-')
plt.title('Simply 20 bins with and without sections')
````

### Using python unpacking list to simplify

The star * operator unpacka a list
```python
myfun(x,p[0],p[1],p[2])
```
can be replaced with 
```python
myfun(x,*p[0:3])
```

### Using sections and step

````python
plt.plot(*tp.smooth.binbyx(x1,y1, sects=2, steps=[0.1,0.4]),'b.-')
plt.plot(*tp.smooth.binbyx(x2,y2, sects=[0,1,2.5,4], steps=[0.05,0.1,0.25]),'r.-')
plt.title('Using sections and steps')
````
